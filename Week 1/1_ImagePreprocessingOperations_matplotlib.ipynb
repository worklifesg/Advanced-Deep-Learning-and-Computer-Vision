```python
import cv2
import numpy
img = cv2.imread('balloon.jpg') #imread to read image
```

```python
type(img)
```

```python
import matplotlib.pyplot as plt
def showImage(img):
    plt.imshow(img)
    plt.title('Cappadocia: Hot Air Balloon Flight')
    plt.show()
```

```python
showImage(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #changing BGR to RGB (convert color option) else image will be in blue channel
showImage(img)
```

```python
img.shape
```

```python
# Rotate the image

img_rotated = img
height,width=img.shape[0:2]

rotationMatrix = cv2.getRotationMatrix2D((width/2,height/2),90,0.5) #Rotation Matrix = 90 rotation and keeping the ratio of 0.5
rotatedImage = cv2.warpAffine(img_rotated,rotationMatrix,(width,height)) #w/2,h/2 is rotation point center

showImage(rotatedImage)
```

```python
# Crop the image

img_crop=img

startRow = int(height*0.15)
endRow = int(height*0.85)

startCol = int(width*0.15)
endCol = int(width*0.85)

croppedImage = img_crop[startRow:endRow,startCol:endCol]

showImage(croppedImage)
```

```python
#Resize the image

img_resize = img

resizedImage = cv2.resize(img_resize,(0,0),fx=0.75,fy=0.75)

showImage(resizedImage)
resizedImage.shape
```

```python
showImage(img)
img.shape
```

```python
#Blur the image (Gaussian Blur)

img_blur = img
blurImage = cv2.GaussianBlur(img_blur,(7,7),0) #multiplying kernel (7,7) with the iamge so it becomes blur and 0 is std

showImage(blurImage)
```

```python
# Blur the image (Median Blur) - mathematical func(mean function)

blurImageMedian = cv2.medianBlur(img_blur,5)#median blur with 50% noise
showImage(blurImageMedian)
```

```python
#Detect Edges in the image

EdgeImage = cv2.Canny(img,100,200) # threshold values in x ,y
showImage(EdgeImage)
```

```python
# Converting color image to gray scale image

grayImage = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
showImage(grayImage)
```
